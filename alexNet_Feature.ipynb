```python
import cv2
import numpy as np
import os
import scipy.io as sio
import torch
import torchvision.transforms as transforms
import torchvision.models as models
import torch.nn as nn
```

```python
def initialize_model():
    # Initialize these variables which will be set in this if statement. Each of these
    #   variables is model specific.
    model_ft = models.alexnet(pretrained=True)
    input_size = 224
    
    return model_ft, input_size
```

```python
# Device configuration
if torch.cuda.is_available():
    print("GPU is active")
    device = torch.device('cuda')
else:
    device = torch.device('cpu')

model, size = initialize_model()

# Taking penultimate Linear Layer from alexNet
model.classifier = nn.Sequential(*list(model.classifier.children())[:-2]) 
model_fc7 = model.to(device)

print(model_fc7)


normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
prep = transforms.Compose([ transforms.ToTensor(), normalize ])
```

```python
inputpath = 'training_set/'

outpath_fc7 = 'alexNet_fc7/'

folder_list = sorted(os.listdir(inputpath))
try:
    folder_list. remove('.DS_Store')
except:
    pass

out_list_fc6 = np.empty(shape=[0,4096])
out_list_fc7 = np.empty(shape=[0,4096])

count = 0
for file_name in sorted(os.listdir(inputpath)):
    img = cv2.imread(inputpath + file_name)
    count = count + 1

    img = cv2.resize(img, (224, 224))
    img  = prep(img)
    inp = torch.stack([img, img])
    inp = inp.to(device)
    
    outputs_fc7 = model_fc7(inp)
    outputs_fc7 = outputs_fc7.mean(0)
    outputs_fc7 = outputs_fc7.cpu().detach().numpy()

    print(count)
    print("saving the file ", file_name+'.mat')
    sio.savemat(outpath_fc7 + file_name +'.mat', {'features':outputs_fc7}, do_compression=True)
    if count == 40000:
        break
```
