```python
import cv2
import numpy as np
import os
import scipy.io as sio
import torch
import torchvision.transforms as transforms
import torchvision.models as models
import torch.nn as nn
```

```python
def get_all_img_in_tensor(img, prep):
    img_list = []
    img1 = img[:224, :224]
    #print(img1.shape)
    img_list.append(prep(img1))
    img2 = img[:224, 151:]
    #print(img2.shape)
    img_list.append(prep(img2))
    img3 = img[138:362, :224]
    #print(img3.shape)
    img_list.append(prep(img3))
    img4 = img[138:362, 151:]
    #print(img4.shape)
    img_list.append(prep(img4))
    img5 = img[276:, :224]
    #print(img5.shape)
    img_list.append(prep(img5))
    img6 = img[276:, 151:]
    #print(img6.shape)
    img_list.append(prep(img6))
    img7 = img[138:362, 75:299]
    #print(img7.shape)
    img_list.append(prep(img7))
    #out = torch.stack([img1, img2])
    return img_list

def initialize_model():
    # Initialize these variables which will be set in this if statement. Each of these
    #   variables is model specific.
    model_ft = models.vgg16_bn(pretrained=True)
    input_size = 224
    
    return model_ft, input_size
```

```python
# Device configuration
if torch.cuda.is_available():
    print("GPU is active")
    device = torch.device('cuda')
else:
    device = torch.device('cpu')

model, size = initialize_model()
model.classifier = nn.Sequential(*list(model.classifier.children())[:-3])
model = model.to(device)

normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
prep = transforms.Compose([ transforms.ToTensor(), normalize ])
```

```python
def get_img_in_tensor(img, prep):
    img_list = []
    resized_img = cv2.resize(img, (224, 224))
    img_list.append(prep(resized_img))
    img2 = resized_img
    img_list.append(prep(img2))
    return img_list


inputpath = 'all_data/train_photos/'
outpath = 'gen_data/vgg7avg/'
#inputpath = 'all_data/test_photos/'
#outpath = 'gen_data/test_data/vgg_resize/'
#outpath = 'gen_data/test_data/vgg7avg/'

folder_list = sorted(os.listdir(inputpath))
try:
    folder_list. remove('.DS_Store')
except:
    pass
#print(folder_list)

i = 1
out_list = np.empty(shape=[0,4096])
for file_name in sorted(os.listdir(inputpath)):
    #print(file_name)
    img = cv2.imread(inputpath + file_name)
    
    #print(img.shape)#[0], img.shape[1], img.shape[2])
    if img.shape != (500,375,3):
        continue

    #print(img.shape)
    #img = img.to(device)
    #img_list = get_img_in_tensor(img, prep)     ## resize
    img_list = get_all_img_in_tensor(img, prep)    ## avg of 7 images
    
    inp = torch.stack(img_list)
    #print(inp.shape)
    inp = inp.to(device)
    #print(out.shape)
    
    outputs = model(inp)
    mean_output = outputs.mean(0)
    np_out = mean_output.cpu().detach().numpy()

    #print("image -", i, mean_output.shape)
    i = i + 1
        
    #out_tensor = torch.stack(out_list)
    #write the out_list to file
    print("saving the file ", file_name+'.vgg')
    sio.savemat(outpath + file_name +'.vgg', {'features':np_out}, do_compression=True)
```
