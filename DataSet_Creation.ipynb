```python
import os
import cv2 
```

```python
"""
Creating the dataset for our project from the original dataset
"""
for filename in os.listdir('train_photos'):
    if filename.startswith("._"): 
        os.remove(os.path.join('train_photos',filename))
        print(os.path.join('train_photos', filename))
        continue
    else:
        continue      
count=0
for filename in os.listdir('train_photos'):
    if filename.endswith(".jpg"):
        count+=1
        continue
    else:
        continue
print(count)   
```

```python
"""
Taking train = 40000
"""
count=0
for filename in sorted(os.listdir('train_photos'),key=lambda x: int(os.path.splitext(x)[0])):
    if filename.endswith(".jpg") and count!=40000: 
        img = cv2.imread(os.path.join('train_photos',filename))
        cv2.imwrite(os.path.join('training_set',filename),img)
        print(os.path.join('training_set', filename))
        count+=1
    elif count == 40000:
        break
print(count)
```

```python
"""
Taking test = 10000
"""
count=0
for filename in sorted(os.listdir('train_photos'),key=lambda x: int(os.path.splitext(x)[0]),reverse = True):
    if filename.endswith(".jpg") and count!=10000: 
        img = cv2.imread(os.path.join('train_photos',filename))
        cv2.imwrite(os.path.join('testing_set',filename),img)
        print(os.path.join('testing_set', filename))
        count+=1
    elif count == 10000:
        break
print(count)
```
